### Encoding the Target Feature

We prepare the dataset for modeling by separating the features from the target variable and encoding the target variable into numerical values. Then, we isolate the target variable by selecting the 'class' column from the dataset and use **LabelEncoder** to convert this categorical target variable into numerical values. ### Feature Selection & Ranking

 Here are the most important 4 features as selected by the **F-score** on the training set. This quick ranking of the top features provides insight into the most relevant factors influencing the problem at hand. We used these selected features for model fitting.

```python
from sklearn.feature_selection import SelectKBest, f_classif
import numpy as np

num_features = 4

fs_fit_fscore = SelectKBest(score_func=f_classif, k=num_features)
fs_fit_fscore.fit_transform(D_train, t_train)

fs_indices_fscore = np.argsort(np.nan_to_num(fs_fit_fscore.scores_))[::-1][0:num_features]

# Retrieve the names and scores of the selected features
best_features_fscore = Data.columns[fs_indices_fscore].values
feature_importances_fscore = fs_fit_fscore.scores_[fs_indices_fscore]


print("Selected Features by F-Score:")
for feature, score in zip(best_features_fscore, feature_importances_fscore):
    print(f"{feature}: {score:.4f}")
```

### Algorithm 1: KNN Classifier

Here, we will consider K values [1, 5, 10, 15, 20, 25, 30] and distance metrics of 1 (Manhattan), 2 (Euclidean)

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV

cv_method_train = StratifiedKFold(n_splits=5, shuffle=True, random_state=999)
```

```python
import numpy as np

params_KNN = {'n_neighbors': [1, 5, 10, 15, 20, 25, 30], 
              'p': [1, 2]}
```

### Algorithm 4: Random Forest

The Random Forest Classifier operates by constructing a multitude of decision trees during training and outputting the class that is the mode of the classes (classification) of the individual trees. 
It Combines the predictions of multiple decision trees to improve accuracy and control over-fitting. Each tree in the forest is trained on a random subset of the data with replacement, ensuring diversity among the trees. At each split in the decision tree, a random subset of features is considered, which helps in reducing correlation among the trees. By averaging the results of many trees, the Random Forest reduces the risk of overfitting, particularly when compared to individual decision trees. It provides a ranking of feature importance, which can be useful for feature selection and understanding the underlying data structure.
ure.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import PowerTransformer

#Initializes a RandomForestClassifier with a fixed random state (999). This ensures reproducibility of results.
rf_classifier = RandomForestClassifier(random_state=999)

#specifyies the hyperparameters to be tuned and their respective ranges/values
params_RF = {'n_estimators': [100, 250, 500], #Number of trees in the forest.
             'max_depth': [5, 7, 10, 12]} #maximum depth of the trees

gs_RF = GridSearchCV(estimator=rf_classifier, 
                     param_grid=params_RF, 
                     cv=cv_method_train,
                     verbose=1,
                     n_jobs=-2,
                     scoring=scoring_metric)

gs_RF.fit(D_Train_fs, t_train);
```

### Algorithm 5: Gradient Boosting Classifier

Gradient Boosting builds an ensemble of trees sequentially, where each tree attempts to correct the errors of the previous one.
This Combines the predictions of multiple weak learners (decision trees) to create a strong learner.  Focuses on improving the accuracy of the model by giving more weight to instances that were misclassified by previous trees : Uses gradient descent to minimize the loss function, which measures the difference between the actual and predicted values.


```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

# Create an instance of GradientBoostingClassifier
gb_classifier = GradientBoostingClassifier(random_state=999)

# Define the parameter grid for Gradient Boosting
params_gb = {
    'n_estimators': [50,100],  # Number of boosting stages
    'learning_rate': [0.1, 0.2],  # Learning rate shrinks the contribution of each tree
    'max_depth': [3, 5]  # Maximum depth of the individual regression estimators
}

# Perform grid search
gs_gb = GridSearchCV(estimator=gb_classifier,
                     param_grid=params_gb,
                     cv=cv_method,
                     verbose=1,
                     n_jobs=-1,
                     scoring=scoring_metric)

# Train the model using the training sets
gs_gb.fit(D_Train_fs, t_train)
```

The output message indicates that GridSearchCV is performing a comprehensive search for the best hyperparameters for the GradientBoostingClassifier using 5-fold cross-validation. The process involves evaluating 8 different hyperparameter combinations, with a total of 40 model fits. This ensures that the model is thoroughly evaluated and the best hyperparameters are selected to optimize performance.

```python
gs_gb.best_params_
```

Above result indicates that the best-performing Gradient Boosting model has a learning rate of 0.1, a maximum tree depth of 3, and consists of 50 trees


```python
gs_gb.best_score_
```

The gs_gb.best_score_ value of 95.20% represents the highest mean cross-validation score achieved during the grid search process for the GradientBoostingClassifier

```python
results_gb = pd.DataFrame(gs_gb.cv_results_)

# Create a line chart for the mean CV scores
plt.figure(figsize=(12, 8))

# Loop over unique learning rates
for lr in results_gb['param_learning_rate'].unique():
    temp = results_gb[results_gb['param_learning_rate'] == lr]
    # Plot for each value of max_depth
    for md in temp['param_max_depth'].unique():
        temp_md = temp[temp['param_max_depth'] == md]
        plt.plot(temp_md['param_n_estimators'], temp_md['mean_test_score'], marker='o', label=f'Learning Rate: {lr}, Max Depth: {md}')

plt.legend()
plt.xlabel('Number of Estimators')
plt.ylabel('Mean CV Score')
plt.title('Gradient Boosting Performance Comparison')
plt.grid(True)
plt.show()
```

The best performance is observed for learning_rate = 0.1 and max_depth = 3 with n_estimators = 50, yielding the highest mean CV score of approximately 0.952.
Increasing the number of estimators to 100 decreases the performance for this configuration, suggesting potential overfitting. This indicates that a smaller learning rate with a shallower tree depth and fewer trees provides the best performance for the given dataset.
The scores for other configurations (learning_rate = 0.1, max_depth = 5 and learning_rate = 0.2) are relatively stable across different values of n_estimators, indicating that these configurations are less sensitive to the number of trees. The plot indicates that the model's performance is sensitive to the number of estimators for certain combinations


## Model Comparison

Model comparison is a critical step in the machine learning workflow that involves evaluating and contrasting the performance of different models to identify the one that best addresses a particular problem. This process ensures that the chosen model generalizes well to unseen data and meets the desired performance criteria. Multiple machine learning models are trained using the training dataset. Test data provides an unbiased evaluation of a model's performance on unseen data. By evaluating models on test data, we can detect overfitting and choose models that perform well on both training and test data.

```python
from sklearn.model_selection import cross_val_score
cv_method_test = StratifiedKFold(n_splits=5, shuffle=True, random_state=999)
```

```python
cv_results_KNN = cross_val_score(estimator=gs_KNN.best_estimator_,
                                 X=D_Test_fs,
                                 y=t_test, 
                                 cv=cv_method_test, 
                                 scoring='roc_auc')
cv_results_KNN.mean()
```

The ROC AUC score measures the ability of the model to distinguish between positive and negative classes. A higher ROC AUC score indicates better discrimination between classes. Therefore, a score of 0.937 suggests that the KNN model performs quite well on the test data, demonstrating good predictive ability in distinguishing between classes.

```python
cv_results_DT = cross_val_score(estimator=gs_DT.best_estimator_,
                                 X=D_Test_fs,
                                 y=t_test, 
                                 cv=cv_method_test, 
                                 scoring='roc_auc')
cv_results_DT.mean()
```

A higher ROC AUC score indicates better discrimination between classes. Therefore, a score of 0.932 suggests that the decision tree model performs quite well on the test data.

```python

D_Test_fs_transformed = PowerTransformer().fit_transform(D_Test_fs)

cv_results_NB = cross_val_score(estimator=gs_NB.best_estimator_,
                                 X=D_Test_fs_transformed,
                                 y=t_test, 
                                 cv=cv_method_test, 
                                 scoring='roc_auc')
cv_results_NB.mean()
```

NaiveBayes classifier provides a score of 0.926, indicating good performance. 

```python
cv_results_RF = cross_val_score(estimator=gs_RF.best_estimator_,
                                 X=D_Test_fs,
                                 y=t_test, 
                                 cv=cv_method_test, 
                                 n_jobs=-2,
                                 scoring='roc_auc')
cv_results_RF.mean()
```

RandomForest provided a higher score of 0.94 in comparison to Naive Bayes and KNN classifier 

```python
gbc = GradientBoostingClassifier()
cv_results_GBC = cross_val_score(estimator=gbc,
                                 X=D_Test_fs,
                                 y=t_test, 
                                 cv=cv_method_test, 
                                 scoring='roc_auc')

cv_results_GBC.mean()
```

```python
from scipy import stats

# any p-value < 0.05 indicates a statistically significant result

print(stats.ttest_rel(cv_results_GBC, cv_results_KNN))
print(stats.ttest_rel(cv_results_GBC, cv_results_DT))
print(stats.ttest_rel(cv_results_GBC, cv_results_RF))
print(stats.ttest_rel(cv_results_GBC, cv_results_NB))
```

All the p-values from the t-tests are greater than the significance level of 0.05. This means that, for each comparison, there is no statistically significant difference in the performance between the GradientBoostingClassifier model and each of the other models (KNN, DT, RF, and NB). The differences in their mean ROC AUC scores are not statistically significant, suggesting that any observed performance differences might be due to random variation in the data rather than true differences in model performance.
